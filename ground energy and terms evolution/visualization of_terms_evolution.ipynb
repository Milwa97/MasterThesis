# Single plots for each series

```python
def plot_single(df):
    
    fig, axs = plt.subplots(3, 1, figsize=(12, 15), sharex=True)

    axs[0].plot(np.array(df.loss)*10, color= 'black', label = 'loss function', linewidth=2)
    axs[0].plot(1-df.term_IV, color= 'indigo', linestyle='--')
    axs[0].plot(1-df.term_IV_star, color= 'plum', linestyle='--')
    axs[0].plot(1-df.term_IV_diamond, color= 'darkviolet', linestyle='--')

    
    axs[0].plot(df.MSE, color= 'indigo', label = '$MSE$',linewidth=2)
    axs[1].plot(df.term_I, color= 'indigo', label = '$T_1$')
    axs[2].plot(df.term_II, color= 'indigo', label = '$T_2$')

    axs[0].plot(df.MSE_star, color= 'plum', label = '$MSE^*$', linewidth=2)
    axs[1].plot(df.term_I_star, color= 'plum', label = '$T_1^*$')
    axs[2].plot(df.term_II_star, color= 'plum', label = '$T_2^*$')

    axs[0].plot(df.MSE_diamond, color= 'darkviolet', label = '$MSE^\diamond$', linewidth=2)
    axs[1].plot(df.term_I_diamond, color= 'darkviolet', label = '$T_1^\diamond$')
    axs[2].plot(df.term_II_diamond, color= 'darkviolet', label = '$T_2^\diamond$')

    for ax in axs:
        ax.grid(True, color='gray', alpha = 0.2)
        ax.legend(fontsize=11)
        ax.tick_params(axis='both', labelsize=11)

    axs[0].set_ylabel('MSE', fontsize=12)
    axs[1].set_ylabel('$T_1$',fontsize=12)
    axs[2].set_ylabel('$T_2$',fontsize=12)
    axs[-1].set_xlabel('epoch',fontsize=12)
    axs[-1].set_xlim(-5,1000)
    
```
